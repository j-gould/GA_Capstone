```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import cv2
```

```python
# 2. define data directory and create train and test datasets
```

```python
# define project dataset directory
dataset =  './dataset/binary/'
```

```python
# define classes
classes = [i for i in os.listdir(dataset) if i[0] != '.']        
print(classes)
```

```python
# create training and validation folders
# https://github.com/jfilter/split-folders
# !pip install split-folders

import split_folders
input_folder = dataset
output = './binary_dataset/'
#split_folders.ratio(input_folder, output = output, seed = 99, ratio = (.8, .2))    # only run once
```

```python
# update project dataset directory
dataset = output
```

```python
# define modelling folders
folders = [i for i in os.listdir(dataset) if i[0] != '.']        
print(folders)
```

```python
# summarise project dataset
dataset_summary = {}
for folder in folders:
    dataset_summary[folder] = {}
    for cl in classes:
        images = os.listdir(dataset + folder + '/' + cl)
        dataset_summary[folder][cl] = len(images)
        
print(dataset_summary)        
```

```python
# 3. inspect sample images
```

```python
for cl in classes:
    cl_ = cl[:2]
    path = dataset + 'train/' + cl + '/' + cl_ + '_1000.jpg'
    print('Image Path:', path)
    original_image = cv2.imread(path)
    resized_image = cv2.resize(original_image, (128,128))
    print('Category:', cl)
    print('Shape of original image:',original_image.shape)
    print('--------------------------------------------------')
    sns.set_style('white')
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.show()
    print('Shape of resized image:',resized_image.shape)
    plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
    plt.show()
```
